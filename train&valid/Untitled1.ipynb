```python
import pickle

# 打开.pkl文件并加载数据
with open('original_valid_set.pkl', 'rb') as file:
    data = pickle.load(file)

# 使用集合来跟踪已经存在的数据
unique_data = set()

# 用于存储去重后的数据
filtered_data = []

# 遍历加载的数据
for item in data:
    # 检查元素是否已经存在
    if item not in unique_data:
        unique_data.add(item)
        filtered_data.append(item)

# 将经过处理的数据重新保存到.pkl文件中
with open('original_valid_set.pkl', 'wb') as file:
    pickle.dump(filtered_data, file)
```
