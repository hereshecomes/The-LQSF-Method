```python
from lib.jianji.code_csf import csf
import torch 
from torch.utils.data import Dataset
from torch import nn
from d2l import torch as d2l
```

```python
traindSet = csf()
trainloader =  torch.utils.data.DataLoader(traindSet, batch_size=256, num_workers=16, drop_last=True, shuffle=True)
testset = csf('validset.pkl')
testloader = torch.utils.data.DataLoader(testset, batch_size=256, num_workers=16, drop_last=True, shuffle=True)
```

```python
net = nn.Sequential(
    nn.Linear(152, 608), nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(608, 1024), nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(1024, 512), nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(512, 128), nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(128, 16), nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(16, 4), nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(4, 1), nn.Sigmoid())
```

```python
for X, y in testloader:
    y_pred = net(X)
    print(y_pred)
    output = torch.where(y_pred>0.5,torch.ones_like(y_pred),torch.zeros_like(y_pred))
    print(output)
    break
```
