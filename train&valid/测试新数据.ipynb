```python
import numpy as np
from keras.models import Model
from keras.layers import Input, LSTM, RepeatVector, Masking, Embedding, Add, Dense, TimeDistributed, Reshape

# 示例数据
data = [
    [3, 3, 1, 1, 3, 1, 1, 3, 3, 1, 1, 1, 2, 1, 3, 2, 1, 4, 1, 2, 2, 1, 4, 1, 4, 1, 4, 3, 1, 2, 3, 2, 4, 2, 2, 1, 4, 4, 1, 2, 1, 1, 2, 2, 3, 3, 2, 2, 3, 2, 3, 2, 2, 4, 1, 1, 1, 2, 3, 1, 1, 4, 4, 2, 2, 1, 4, 1, 4, 1, 4, 3, 2, 2, 4, 2, 4, 1, 4, 1, 4, 4, 2, 4, 4, 3, 2, 4, 4, 2, 4, 1, 2, 2, 3, 1, 4, 3, 1, 4, 4, 1, 1, 3, 1, 1, 2, 1, 4, 2, 4, 1, 3, 2, 3, 4, 4, 3, 1, 2, 4, 1, 3, 2, 2, 1, 2, 3, 2, 4, 4, 3, 1, 3, 2, 2, 4, 2, 1, 4, 3, 4, 4, 4, 1, 2, 2, 3, 3, 3, 1, 3],
    [1, 2, 3, 4, 1, 2, 3]
]

# 设置 max_len 为 152
max_len = 152
data_padded = np.array([seq + [0] * (max_len - len(seq)) for seq in data])

# 定义位置编码
position_indices = np.arange(max_len)
position_encoding = np.tile(position_indices, (len(data_padded), 1))

# 构建模型
input_dim = data_padded.shape[1]
encoding_dim = 50

input_seq = Input(shape=(input_dim,))
position_input = Input(shape=(input_dim,))
embedded_seq = Embedding(input_dim, 10)(input_seq)  # 使用嵌入层处理输入
position_embedding = Embedding(input_dim, 10)(position_input)  # 位置编码的嵌入层
combined_input = Add()([embedded_seq, position_embedding])  # 组合输入和位置编码

masked_input = Masking(mask_value=0)(combined_input)  # Masking layer 处理填充部分
encoded = LSTM(encoding_dim, activation='relu')(masked_input)
decoded = RepeatVector(input_dim)(encoded)
decoded = LSTM(10, return_sequences=True)(decoded)
decoded = TimeDistributed(Dense(1))(decoded)  # 使用 TimeDistributed 包装 Dense 层以逐时间步应用

autoencoder = Model([input_seq, position_input], decoded)
autoencoder.compile(optimizer='adam', loss='mse')

# 训练模型
data_padded_reshaped = data_padded.reshape((data_padded.shape[0], data_padded.shape[1], 1))
autoencoder.fit([data_padded_reshaped, position_encoding], data_padded_reshaped, epochs=100, batch_size=16, shuffle=True)

# 使用编码器部分来获得降维后的数据
encoder = Model([input_seq, position_input], encoded)
encoded_data = encoder.predict([data_padded_reshaped, position_encoding])

print("降维后的数据：", encoded_data)
```
