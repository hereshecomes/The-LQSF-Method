```python
import pandas as pd
import numpy as np
str = f'./test_data/train_1.xlsx'
df1 = pd.read_excel(str)
new_columns = np.arange(152)
df1.columns = new_columns
df1[152]=1
df1
```

```python
import pandas as pd
import numpy as np
str = f'./test_data/train_0.xlsx'
df2 = pd.read_excel(str)
new_columns = np.arange(152)
df2.columns = new_columns
df2[152]=0
df2
```

```python
file_train_dataset = pd.concat([df1, df2], axis=0)
```

```python
file_train_dataset.to_pickle("fountain_seq.pkl")
```

```python
#阴性集上采样(具体乘多少取决于阴阳性集比例)
```

```python
new_df = [df2]*20
all_df = pd.concat(new_df, axis=0)
finaldf = pd.concat([all_df,df1],axis=0)
finaldf
```

```python
finaldf.to_pickle("balanced_fountain_seq.pkl")
```
