```python
import pandas as pd
import numpy as np
df = pd.read_pickle('positive_original.pkl')
df1 = pd.read_pickle('neg_original.pkl')
```

```python
df[152]=1
```

```python
df_final = pd.concat([df, df1], axis=0)
```

```python
def shuffle(seed, dataframe):
    length = len(dataframe)
    data_index = list(range(length))
    np.random.seed(seed)
    np.random.shuffle(data_index)
    df_temp = dataframe.iloc[data_index]
    return df_temp

new_df = shuffle(50, df_final)
```
